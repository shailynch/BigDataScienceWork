```python
#import require libraries
from scipy.stats import beta
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm,chi2,gamma,probplot
```

```python
#a create subplot
fig,a = plt.subplots(2,3)
#array subplots
k = 3
N = [5,10,50]

size = 100000
#variable size 

x2k = []
for n in N:
    params = n*k/2
    #parameters for distribution
    distribution = chi2.rvs(params,size=[size,n])
    x2k.append(distribution)

print("The shape of N=5: ",x2k[0].shape)
print("The shape of N=10:",x2k[1].shape)
print("The shape of N=50:",x2k[2].shape)
```

```python
mean_n5 = np.mean(x2k[0],axis=1)
mean_n10 = np.mean(x2k[1],axis=1)
mean_n50 = np.mean(x2k[2],axis=1)
```

```python
alpha5,loc5,scale5 = gamma.fit(mean_n5)
alpha10,loc10,scale10 = gamma.fit(mean_n10)
alpha50,loc50,scale50 = gamma.fit(mean_n50)
```

```python
pdf_5 = gamma.pdf(mean_n5,alpha5,loc5,scale5)
pdf_10 = gamma.pdf(mean_n10,alpha10,loc10,scale10)
pdf_50 = gamma.pdf(mean_n50,alpha50,loc50,scale50)
```

```python
fig,a = plt.subplots(2,3,figsize=(10,10))
a[0,0].hist(mean_n5,bins=100,density=True,color='r')
a[0,0].scatter(mean_n5,pdf_5)
probplot(mean_n5,alpha5,dist='gamma',plot=a[1,0])
a[0,1].hist(mean_n10,bins=100,density=True,color='g')
a[0,1].scatter(mean_n10,pdf_10)
probplot(mean_n10,alpha10,dist='gamma',plot=a[1,1])
a[0,2].hist(mean_n50,bins=100,density=True,color='y')
a[0,2].scatter(mean_n50,pdf_50)
probplot(mean_n50,alpha50,dist='gamma',plot=a[1,2])
plt.show()
```

# Question 3

```python
sampleSize = 9000
p = 0.1 
observedMean = 945
theoreticalMean = sampleSize * p

#variance calculation
var =sampleSize * p * (1-p)
#standard deviation calculation
standardDev = np.sqrt(var)
#Z score calculation
mean = observedMean - theoreticalMean
zScore = mean/standardDev

pValue = 1 - norm.cdf(zScore)

if pValue < 0.025:
    conclusion = 'Null hypothesis to be rejected'
else:
    conclusion = 'Null hypothesis not to be rejected'

print(f" the p value is {round(pValue,3)},{conclusion} ")
```
