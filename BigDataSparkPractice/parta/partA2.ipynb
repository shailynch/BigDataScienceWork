```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import seaborn as sns
```

```python
#Load the average.txt file
with open('time_analysis_average.txt') as f:
    data = json.load(f)

#Split the data  and convert data to DF, change column type
data = [i.split(',') for i in data]
df = pd.DataFrame(data, columns=["Date", "Average Transaction"])
df['Average Transaction'] = df['Average Transaction'].astype(float)

#Amend and remove day from date then sort 
df['Date'] = pd.to_datetime(df['Date'], format='%m-%Y')
df['Date'] = df['Date'].dt.to_period('M')
df = df.sort_values(by='Date', ascending=True)
```

```python
df.head(10)
```

```python
#Barplot of the findings belw
plt.figure(figsize=(20,10))
sns.barplot(x="Date", y="Average Transaction", data=df,  palette=["red", "orange", "green"])
plt.title("Average Transactio per monnth")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.ylabel("Average Transaction per month")
plt.show()
```
