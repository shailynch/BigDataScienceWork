```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import seaborn as sns
```

```python
#Load the total.txt file into DF
with open('time_analysis_total.txt') as f:
    data = json.load(f)
df = pd.DataFrame(data, columns=["Date", "Transactions"])

# alter date format and sort 
df['Date'] = pd.to_datetime(df['Date'], format='%m-%Y')
df['Date'] = df['Date'].dt.to_period('M')
df = df.sort_values(by='Date', ascending=True)
```

```python
df.head(10)
```

```python
#PLot a bar graph of the monthly transactions 
plt.figure(figsize=(20,10))
sns.barplot(x="Date", y="Transactions", data=df, palette=["red", "orange", "green"])
plt.title("Total Transactions per Month")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.ylabel("Transactions per Month")
plt.show()
```
