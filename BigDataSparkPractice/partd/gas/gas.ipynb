# Gas Price 

```python
import pandas as pd
import json 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
#Load the price.txt file 
with open('price.txt') as f:
    data = json.load(f)

#Create a DataFrame from the data
df = pd.DataFrame(data, columns=["Date","Gas Price"])

#Change the  Gas Price to float 
df["Gas Price"] = df["Gas Price"].astype(float)

#Change the Date to datetime aand sort 
df["Date"] = pd.to_datetime(df["Date"], format="%m-%Y")
df.sort_values(by=['Date'], inplace=True, ascending=True)
```

```python
#Create a line plot of the price 
plt.figure(figsize=(20,10))
plt.title("Gas Price", fontsize=18)
plt.plot(df["Date"], df["Gas Price"], color="green")
plt.xlabel("Date")
plt.xticks(df["Date"], rotation=45)
plt.ylabel("Gas Price")
plt.show()
```

# Gas Usage 

```python
#load the usage.txt file 
with open('usage.txt') as f:
    data = json.load(f)
    
#Create a DataFrame from the data
gas_df = pd.DataFrame(data, columns=["Date","Gas Usage"])

#Change the  Gas Usage to float
gas_df["Gas Usage"] = gas_df["Gas Usage"].astype(float)

#Change the Date to datetime
gas_df["Date"] = pd.to_datetime(gas_df["Date"], format="%m-%Y")


#Sort the value by date in an ascending order
gas_df.sort_values(by="Date", inplace=True)
```

```python
#Create a line plot of the change in gas Usage 
plt.figure(figsize=(20,10))
plt.title("Gas Usage", fontsize=18)
plt.plot(gas_df["Date"], gas_df["Gas Usage"], color='green')
plt.xlabel("Date")
plt.xticks(gas_df["Date"], rotation=45)
plt.ylabel("Gas Usage")
plt.show()
```
